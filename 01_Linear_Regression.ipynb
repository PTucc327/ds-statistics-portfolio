```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

# 1. Load the dataset object
boston_data = load_boston()

# 2. Create the DataFrame using the feature data and column names
# The data is in a dictionary-like structure, access 'data' for features
# and 'feature_names' for column labels.
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)

# 3. Add the target variable (MEDV - Median value) to the DataFrame
df['MEDV'] = boston_data.target

# 4. Display the first few rows to verify the load
print("--- Python DataFrame Head ---")
print(df.head())

# The 'CHAS' column (Charles River dummy) is now ready for your T-test!
# The 'RM' and 'MEDV' columns are ready for your Regression!
```
