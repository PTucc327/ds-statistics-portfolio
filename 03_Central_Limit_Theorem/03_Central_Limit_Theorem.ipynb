## What is the Central Limit Theorem?
The Central Limit Theorem states that if you take a large enough random samples from any population, the distribution of the sample means would reflect a normal distribution (bell curve), even if the full populations distribution is skewed

### Step 1: Import necessary libraries and data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
population = df['median_house_value']
```

### Step 2: Show the Original Distribution of data

```python
plt.figure( figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(population, bins=30, kde=True)
plt.title('Population Distribution of Median House Values')
```

From the original plot it can be seen that the distribution is not a normal distribution and is depicting a right skewed distribution.

### Step 3: Simulate Central Limit Theorem

```python
sample_sizes = [5, 30, 100]
plt.figure(figsize=(15, 5))
```

```python
sample_sizes = [5, 30, 100]
plt.figure(figsize=(15, 5))

for i, n in enumerate(sample_sizes):
    # Take 1000 samples for each size
    sample_means = [df['median_income'].sample(n, replace=True).mean() for _ in range(1000)]    
    # Create subplots
    plt.subplot(1, 3, i+1)
    sns.histplot(sample_means, kde=True, color='teal')
    plt.title(f'Sample Size n = {n}')
    plt.xlabel('Sample Mean Income')

plt.tight_layout()
plt.show()
```

Looking at the graphs above, it is clear that as the sample size increases, the change in the distribution changes as well. As n approaches 100 it can be noted that the distribution is now a normal distribution.

## What can this tell us?

Using a smaller sample size like 5 in our example can be too small and show a very skewed portion of the data leading to insights that may not necessarily be true about our data. Looking at larger sample sizes like n = 100 can lead to a better assessment and a more precise estimate of the population mean.
